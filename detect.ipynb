```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('NaHeeDo.jpg')
img
```

```python
plt.imshow(img)
```

```python
conv_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(conv_img)
```

```python
face_cas=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
face_cas
```

```python
faces=face_cas.detectMultiScale(conv_img,minNeighbors=19)
faces
```

```python
for (x,y,w,h) in faces:   #137 ,75, 69, 69
    cv2.rectangle(conv_img,(x,y),(x+w,y+h),(0,0,255),4)
```

```python
plt.imshow(conv_img)
```

```python
video=cv2.VideoCapture('video.mp4')
while True:
    success,img=video.read()
    if success:
        gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        faces=face_cas.detectMultiScale(gray,minNeighbors=10)
        for (x,y,w,h) in faces:
            cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),3)
        cv2.imshow("IMAGE",img)
        if cv2.waitKey(1) & 0XFF==ord('q'):
            break
    else:
        print("Video finished")
        break
video.release()
cv2.destroyAllWindows()
```

```python
video=cv2.VideoCapture(0)
while True:
    success,img=video.read()
    if success:
        gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        faces=face_cas.detectMultiScale(gray,minNeighbors=10)
        for (x,y,w,h) in faces:
            cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,0),3)
        cv2.imshow("IMAGE",img)
        if cv2.waitKey(1) & 0XFF==ord('q'):
            break
video.release()
cv2.destroyAllWindows()
```
